```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
UC1 = pd.read_csv("/content/IT304S/UC1.csv")
```

```
UC1.DEMANDA_REGISTRADA_P.plot()
```

$y=A+B \log x$

```
coefficients  = np.polyfit(np.log(range(1,70,1)), UC1.DEMANDA_REGISTRADA_P, 1)
def logcurve(x, coefficients):
  A = coefficients[1]
  B = coefficients[0]
  y = A + (np.log(x)*B)
  return y
x = range(1,70,1)
plt.plot(x, logcurve(x, coefficients),"--", label="fit")
UC1.DEMANDA_REGISTRADA_P.plot()

plt.legend(['Ajuste Log', 'DEMANDA_REGISTRADA_P'])
```
